```python
x = np.linspace(0.05, 10, 1000)
y = np.cos(x)
plt.plot(x, y, ls="-", lw=2, label="plt figure")
plt.legend()
plt.show()
```

```python
x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)
plt.scatter(x, y, label="scatter figure")
plt.legend()
plt.show()
```

```python
x = np.linspace(0.0, 10, 1000)
y = np.random.rand(1000)
plt.scatter(x, y, label="scatter figure")
plt.legend()
plt.xlim(-0.5, 10.5)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)
plt.plot(x, y, ls="-.", lw=2, c="g", label="plot figure")
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
#plt.xlim(-0.5, 10.5)
#plt.ylim(-0.1, 1.1)
plt.show()
```

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)
plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(linestyle=":", color="r")
#plt.xlim(-0.5, 10.5)
#plt.ylim(-0.1, 1.1)
plt.show()
```

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)
plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()
plt.axhline(y=0.0, c="r", ls="--", lw=2)
plt.axvline(x=4.0, c="r", ls="--", lw=2)
#plt.xlabel("x-axis")
#plt.ylabel("y-axis")
#plt.xlim(-0.5, 10.5)
#plt.ylim(-0.1, 1.1)
plt.show()
```

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)
plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()
plt.axvspan(xmin=4.0, xmax=6.0, facecolor="y", alpha=0.3)
plt.axhspan(ymin=0.0, ymax=0.5, facecolor="y", alpha=0.3)
#plt.xlabel("x-axis")
#plt.ylabel("y-axis")
#plt.xlim(-0.5, 10.5)
#plt.ylim(-0.1, 1.1)
plt.show()
```
